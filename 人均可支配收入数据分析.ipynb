```python
# 库导入
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 加载csv文件
data = pd.read_csv('人均可支配收入.csv')
data
```

```python
x = []
y = []
start_row = 30
for index, row in data.iloc[start_row:].iterrows():
    x.append(int(row['年份']))
    y.append(float(row['人均可支配收入']))
x[0], x[len(x)-1]
```

```python
# 画图

# 指定默认字体
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei'] 

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'o-') # o-表示用圆圈标记数据点，-表示用线连接数据点
plt.xlabel('年份')
plt.ylabel('人均可支配收入')
plt.title('1978-2023年人均可支配收入')
plt.show()
```
