```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#加载csv文件
data = pd.read_csv('商品建筑销售价格年初至今平均住宅.csv')
#获取数据
data
```

```python
x = [] #存放日期
y = [] #存放价格
start_row = 28 #开始行数
for index, row in data.iloc[start_row:].iterrows():
    x.append(index - start_row)
    y.append(float(row[1]))
len(y)
```

```python
#解决中文显示问题
plt.rcParams['font.sans-serif'] = ['SimHei']
#创建画布
plt.figure(figsize=(10,6))
#plt.ylim(0, 1000)
#绘制折线图
plt.plot(x,y)
#设置标题
plt.title('商品住宅销售价格指数')
#设置x轴标签
plt.xlabel('时间')
#设置y轴标签
plt.ylabel('商品住宅销售价格指数')
#显示图形
plt.show()
```

```python
# 移动平均法函数定义
def moving_average(data, window_size):
    window = np.ones(int(window_size))/float(window_size)
    return np.convolve(data, window, 'valid')
```

```python
# 移动平均计算
window_size = 24
moving_avg = moving_average(y, window_size)
moving_avg.shape
```

```python
# 作图可视化
plt.figure(figsize=(10,6))
plt.plot(x, y, label='原始数据')
plt.plot(x[int(window_size/2)-1:len(x)-int(window_size/2)], moving_avg, label='移动平均数据')
plt.title('商品住宅销售价格指数')
plt.xlabel('时间')
plt.ylabel('商品住宅销售价格指数')
plt.legend()
plt.show()
```

```python
# 移动平均趋势剔除
detrended = [y[i+int(window_size/2)]/moving_avg[i] for i in range(0, len(moving_avg))]
# 作图可视化
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei'] # 指定默认字体,以支持负号显示
plt.figure(figsize=(10,6))
plt.plot(x[0:len(detrended)], detrended)
plt.axhline(1, color='orange', linestyle = '--', lw=1)
plt.title('季节变动比')
plt.xlabel('时间')
plt.ylabel('商品住宅销售价格指数/移动平均数据')
plt.show()
```
