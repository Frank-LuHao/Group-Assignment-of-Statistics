```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# 数据加载
all_populations = pd.read_csv('人口.csv')
urban_rate  = pd.read_csv('人口数城镇化率常住.csv')
# 数据显示
all_populations, urban_rate
```

```python
# 读取人口
x = []
z = []
y_populations = []
y_urban_rate = []
start_row_population = 61
start_row_urban_rate = 28
for index, row in all_populations.iloc[start_row_population:].iterrows():
    x.append(int(row[0]))
    y_populations.append(float(row[1]))
for index, row in urban_rate.iloc[start_row_urban_rate:].iterrows():
    y_urban_rate.append(float(row[1]))

x
```

```python
urban_populations = [a*b/100 for a, b in zip(y_populations, y_urban_rate)]
urban_populations
```

```python
# 作图
# 中文显示
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
# 创建画布
fig = plt.figure(figsize=(10, 6))
# 添加标题
plt.title('2010-2019年中国城镇人口数')
# 添加x轴标签
plt.xlabel('年份')
# 添加y轴标签
plt.ylabel('人口数（百万人）')
# 显示
plt.plot(x, y_populations, label='总人口数')
plt.plot(x, urban_populations, label='城镇人口数')
plt.legend() # 显示图例
plt.show()
```

```python
# 可视化urban_rate
# 创建画布
fig = plt.figure(figsize=(10, 6))
# 添加标题
plt.title('1982-2023年中国城镇化率')
# 添加x轴标签
plt.xlabel('年份')
# 添加y轴标签
plt.ylabel('城镇化率')
# 显示
plt.plot(x, y_urban_rate, '-o', label='城镇化率')
plt.legend() # 显示图例
plt.show()
```
