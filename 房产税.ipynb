```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
house_tax = pd.read_csv('财政收入一般公共预算收入税收房产税.csv')
house_tax
```

```python
x = []
y = []
start_row = 28
for index, row in house_tax.iloc[start_row:].iterrows():
    x.append(int(row[0]))
    y.append(float(row[1]))
x, y
```

```python
# 画图
# 中文显示
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
# 作图
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'o-', label='房产税')
plt.xlabel('年份')
plt.ylabel('亿元')
plt.title('房产税')
plt.legend()
plt.show()
```
