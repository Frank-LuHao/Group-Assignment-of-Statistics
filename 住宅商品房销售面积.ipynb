```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# 加载 csv 文件
data = pd.read_csv('房屋销售面积35个城市住宅.csv')
# 读取数据
data
```

```python
x = []
y = []
start_row = 28
for index, row in data.iloc[start_row:].iterrows():
    x.append(int(row[0]))
    y.append(float(row[1]))
x,y
```

```python
# 画图
# 解决中文显示
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
# 创建画布
plt.figure(figsize=(10, 6))
# 绘制折线图
plt.plot(x, y, '-o')
# 添加标题
plt.title('房屋销售面积累计35个城市住宅')
# 添加横坐标标签
plt.xlabel('时间')
# 添加纵坐标标签
plt.ylabel('销售面积')
# 显示
plt.show()
```
